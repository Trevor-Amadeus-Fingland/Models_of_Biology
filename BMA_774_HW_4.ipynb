```python
L=5000
N=1000
D=1.10 #These constants will be used for all of 1
dx=L/(N-1)
dt=1
A=50
D_dt_dx2=(D*dt)/(dx**2)
Grid_1=np.zeros((N,501))
Grid_1[N-1,:]=A
#IC
for m in range(len(Grid_1[:,0])):
    if m <= (N-1)/2:
        Grid_1[m][0]=0
    else:
        Grid_1[m][0]=A
plt.figure(figsize=(12,6))
#Actual iteration
for v in range(len(Grid_1[0,:])-1):
    for g in range(len(Grid_1[:,0])):
        if g == 0:
            continue
        if g == N-1:
            continue
        Grid_1[g][v+1]=Grid_1[g][v]+D_dt_dx2*(Grid_1[g+1][v]-2*Grid_1[g][v]+Grid_1[g-1][v])

for l in range(6):
    plt.plot(Grid_1[:,l*100],label=('Time='+str(dt*l*100)))
    
plt.xlim(480,520)
plt.legend()
plt.xlabel('length')
plt.ylabel('concentration')
```

### Question 3(f)

A key thing in using the differencing methods on first order pdes is the necessity to make a decision on the BC.  The decision on which boundary to use is dependent upon which differencing scheme used.  Backwards require that the edge of your grid be defined in order for the iterations to work and the opposite is true for the forwards.  In addition, it is essential to note that not all differencing methods are equivalent as evidenced by question 3. If in research a given differencing method fails, it may be worth trying an alternative to see if it is due to instability of the other method.
Finally, in the case of these advections with non-smooth shape, differencing won't totally replicate the shape of your initial wave due to the infinite slope of the square shaped wave.  However, this can be alleviated by choosing small enough dx.
